# Объектно-Ориентированное Программирование (ООП)

В истории программирования принято считать, что первую в мире программу написала английская графиня Ада Августа Лавлейс. В 1843 году она реализовала алгоритм, решающий уравнение Бернулли, данное уравнение выражает закон сохранения энергии движущейся жидкости. Получается что первая программа была призвана решать сугубо математическую задачу и это неудивительно, ведь именно у физиков и математиков впервые возникла потребность в компьютерных вычислениях. Существуют задачи, которые требуют долгих итерационных вычислений, разумеется людям хотелось хоть как-то упростить этот процесс. Ведь решая уравнение или систему уравнений, человек следует определенным инструкциям, так почему-бы не создать машину, которую можно было бы запрограммировать на выполнение данных инструкций? Именно этим вопросом великие умы задавались еще до начала нашей эры. 

Первые программы выглядели пугающе для обывателей, даже более пугающе чем сейчас. Ведь в начале 1950-ых код программы состоял только из чисел: адресов ячеек памяти и номеров операций. Например, чтобы сложить переменные a и b, надо было знать адреса этих переменных и номер операции сложения, таким образом код выглядел примерно так: 22 01 49. Но не будем вдаваться в историю, а подчеркнем тот факт, что изначально программирование было инструментом ученых и инженеров, а написание и отладка программ являлись кропотливым трудом. 

Проще говоря, первые программы представляли из себя функции (в математическом понимании этого слова). Они принимали на вход начальные условия системы (набор чисел) и выдавали решение (тоже набор чисел). Даже первый высокоуровневый язык назывался Fortran - FORmula TRANslator. Но чем больше программирование набирало популярность, тем большее количество людей трудились над тем, чтобы упростить процесс написания программ. Ученые хотели хранить несколько функций и переменных в рамках одной задачи, создавать копии задач и осуществлять взаимодействие между ними.
