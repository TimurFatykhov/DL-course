```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sinfunc(x):
    y = np.sin(22*(-x)) * 0.2 * x** 2
    return  y
```

```python
x = np.linspace(-6, 6, 10000)
y = x ** 2 - 5
```

```python
fsize = 25

fig = plt.figure(figsize=(13, 10))
plt.grid()
plt.xlim(-5.9, 5.9)
plt.ylim(-5.9, 5.9)
plt.hlines(0, -6, 6, linewidth=1)
plt.vlines(0, -20, 20, linewidth=1)
plt.xticks(np.arange(-5, 6), fontsize=fsize)
plt.yticks(np.arange(-5, 6), fontsize=fsize)

# sin(22*(-x)) * 0.2 * x^2
plt.plot(x, y, color='blue', linewidth=4, alpha=1, label='x^2 - 5', zorder=10)

points = np.array([0.75])
plt.scatter([points], [(points)**2 - 5], s=300, color='red', zorder=20)

# plt.plot(-x, -y, color='red', linewidth=5, label='1/x')
plt.legend(fontsize=fsize)
```

```python
x_l = np.linspace(-6, -0.01, 10000)
x_r = np.linspace(0.001, 6, 10000)
y_l = 1/ x_l
y_r = 1/ x_r
```

```python
fsize = 25

fig = plt.figure(figsize=(13, 10))
plt.grid()
plt.xlim(-5.9, 5.9)
plt.ylim(-5.9, 5.9)
plt.hlines(0, -6, 6, linewidth=1)
plt.vlines(0, -20, 20, linewidth=1)
plt.xticks(np.arange(-5, 6), fontsize=fsize)
plt.yticks(np.arange(-5, 6), fontsize=fsize)

plt.plot(x_l, y_l, color='blue', linewidth=4, zorder=10)
plt.plot(x_r, y_r, color='blue', linewidth=4, label='1/x', zorder=10)

points = np.array([-1])
plt.scatter([points], [1/(points)], s=300, color='red', zorder=20)

# plt.plot(-x, -y, color='red', linewidth=5, label='1/x')
plt.legend(fontsize=fsize)
```
