```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

–ö–∞–∫ –º—ã –≤—ã—è—Å–Ω—è–ª–∏, —Ü–≤–µ—Ç–Ω—ã–µ –∫–∞—Ä—Ç–∏–Ω–∫–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç—Å—è –≤ –≤–∏–¥–µ —Ç—Ä–µ—Ö –º–∞—Ç—Ä–∏—Ü (RGB - –ø–æ –æ–¥–Ω–æ–π –Ω–∞ –∫–∞–∂–¥—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É —Ü–≤–µ—Ç–∞: red, green, blue), –ø–æ—ç—Ç–æ–º—É —Ñ—Ä–∞–∑–∞ "—Å—á–∏—Ç–∞—Ç—å –∫–∞—Ä—Ç–∏–Ω–∫—É" –æ–∑–Ω–∞—á–∞–µ—Ç –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –µ–µ –≤ –≤–∏–¥–µ —Ç–µ–Ω–∑–æ—Ä–∞, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π —è–∑—ã–∫–∞ Python (–Ω–∞–ø—Ä–∏–º–µ—Ä, —ç—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å `numpy.ndarray`). –§—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–ø–æ–ª–Ω—è–µ—Ç —ç—Ç—É –∑–∞–¥–∞—á—É: `cv2.imread()` - –æ–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø—É—Ç—å –¥–æ –∫–∞—Ä—Ç–∏–Ω–∫–∏, –∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ç–µ–Ω–∑–æ—Ä

```python
img = cv2.imread('./images/raccoon.jpg') # –∑–∞–ø–∏—à–µ–º –∫–∞—Ä—Ç–∏–Ω–∫—É –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é img
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫–æ–π –∂–µ —Ç–∏–ø —É –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π img?

```python
type(img)
```

–î–∞ —ç—Ç–æ –∂–µ –∑–Ω–∞–∫–æ–º—ã–π –Ω–∞–º numpy.ndarray! –≠—Ç–æ –æ–∑–Ω–∞—á–µ—Ç, —á—Ç–æ —Ç–µ–ø–µ—Ä—å –º—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–µ—Å—å —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ numpy –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º –∏ –≤ –≥–ª–∞–≤–µ [Numpy-–æ–ø–µ—Ä–∞—Ü–∏–∏ –Ω–∞–¥ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏](#h2) –º—ã –ø–æ–≥–æ–≤–æ—Ä–∏–º –æ–± —ç—Ç–æ–º –ø–æ–¥—Ä–æ–±–Ω–µ–µ. –ê –ø–æ–∫–∞ –¥–∞–≤–∞–π—Ç–µ —Ö–æ—Ç—è –±—ã –≤—ã–≤–µ–¥–µ–º –∑–Ω–∞–∫–æ–º–æ–µ –Ω–∞–º —Å–≤–æ–π—Å—Ç–≤–æ `shape` –∏ —É–∑–Ω–∞–µ–º —Ñ–æ—Ä–º—É –Ω–∞—à–µ–≥–æ —Ç–µ–Ω–∑–æ—Ä–∞

```python
img.shape
```

–ú—ã –≤–∏–¥–∏–º, —á—Ç–æ –≤—ã—Å–æ—Ç–∞ –∫–∞—Ä—Ç–∏–Ω–∫–∏ (–∫–æ–ª-–≤–æ –ø–∏–∫—Å–µ–ª–µ–π –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª–∏) —Ä–∞–≤–Ω–∞ 400, —à–∏—Ä–∏–Ω–∞ (–∫–æ–ª-–≤–æ –ø–∏–∫—Å–µ–ª–µ–π –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏) 640, –∏ 3 –∫–∞–Ω–∞–ª–∞ –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã —Ü–≤–µ—Ç–∞ (–µ—â–µ —ç—Ç–æ –Ω–∞–∑—ã–≤–∞—é—Ç –≥–ª—É–±–∏–Ω–æ–π –∫–∞—Ä—Ç–∏–Ω–∫–∏).

–¢–µ–ø–µ—Ä—å –¥–∞–≤–∞–π—Ç–µ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –Ω–∞—à—É –∫–∞—Ä—Ç–∏–Ω–∫—É üëÄ. –î–ª—è —ç—Ç–æ–≥–æ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π `plt.imshow()` (imshow - —Å–æ–∫—Ä–∞—â–µ–Ω–∏–µ –æ—Ç "image show").

```python
plt.imshow(img) # –Ω–∞ –≤—Ö–æ–¥ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ–¥–∞–µ–º —Ç–µ–Ω–∑–æ—Ä
```

–ï–Ω–æ—Ç –≤ –ø–æ–∑–¥–Ω–µ–µ –≤—Ä–µ–º—è —Å—É—Ç–æ–∫ –ø—Ä–∏ —Ö–æ–ª–æ–¥–Ω–æ–º –æ—Å–≤–µ—â–µ–Ω–∏–∏? –ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —Ü–≤–µ—Ç–∞ –≤ –∫–∞—Ä—Ç–∏–Ω–∫–µ –¥–æ–ª–∂–Ω—ã –≤—ã–≥–ª—è–¥–µ—Ç—å –∏–Ω–∞—á–µ. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é, openCV –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç, —á—Ç–æ –∫–∞–Ω–∞–ª—ã –≤ –∫–∞—Ä—Ç–∏–Ω–∫–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã –≤ –ø–æ—Ä—è–¥–∫–µ Blue Green Red, –ø–æ—ç—Ç–æ–º—É –∫–∞–Ω–∞–ª—ã –≤ –Ω–∞—à–µ–π –∫–∞—Ä—Ç–∏–Ω–∫–µ –ø–µ—Ä–µ–ø—É—Ç–∞–Ω—ã (–∫—Ä–∞—Å–Ω—ã–π –Ω–∞ –º–µ—Å—Ç–µ —Å–∏–Ω–µ–≥–æ, –∞ —Å–∏–Ω–∏–π –Ω–∞ –º–µ—Å—Ç–µ –∫—Ä–∞—Å–Ω–æ–≥–æ) –∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤—ã–≥–ª—è–¥–∏—Ç –Ω–µ –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω–æ. –ß—Ç–æ–±—ã –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–∞—Ä—Ç–∏–Ω–∫—É –∏–∑ BGR –≤ RGB, –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π `cv2.cvtColor()` (convert color). –°–µ–π—á–∞—Å –Ω–∞–º –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è –ª–∏—à—å –ø–µ—Ä–µ—Å—Ç–∞–≤–∏—Ç—å –∫–∞–Ω–∞–ª—ã –º–µ—Å—Ç–∞–º–∏, –Ω–æ —ç—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è —Å–ø–æ—Å–æ–±–Ω–∞ —Ç–∞–∫–∂–µ –æ—Ç–æ–±—Ä–∞–∂–∞—Ç—å –∫–∞—Ä—Ç–∏–Ω–∫—É –≤ –æ—Ç–ª–∏—á–Ω—ã–µ –æ—Ç RGB —Ü–≤–µ—Ç–æ–≤—ã–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ [HSV](https://ru.wikipedia.org/wiki/HSV_(%D1%86%D0%B2%D0%B5%D1%82%D0%BE%D0%B2%D0%B0%D1%8F_%D0%BC%D0%BE%D0%B4%D0%B5%D0%BB%D1%8C)), [LAB](https://ru.wikipedia.org/wiki/LAB), [HSL](https://ru.wikipedia.org/wiki/HSL) –∏ [–¥—Ä—É–≥–∏–µ](https://ru.wikipedia.org/wiki/%D0%A6%D0%B2%D0%B5%D1%82%D0%BE%D0%B2%D0%BE%D0%B5_%D0%BF%D1%80%D0%BE%D1%81%D1%82%D1%80%D0%B0%D0%BD%D1%81%D1%82%D0%B2%D0%BE). –ê–ª–≥–æ—Ä–∏—Ç–º –∫–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏–∏ –∑–∞–¥–∞–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è. –ß—Ç–æ–±—ã –Ω–µ –∑–∞–ø–æ–º–∏–Ω–∞—Ç—å –∫–∞–∫–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º –∫–∞–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω, opencv —Ö—Ä–∞–Ω–∏—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏ –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–∞–∂–¥–æ–π –∏–∑ –∫–æ–Ω—Å—Ç–∞–Ω—Ç –æ—Ç—Ä–∞–∂–∞–µ—Ç –¥–µ–π—Å—Ç–≤–∏–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞, –≤–æ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø—Ä–∏–º–µ—Ä–æ–≤:

```python
cv2.COLOR_BGR2RGB, cv2.COLOR_RGB2LAB, cv2.COLOR_RGB2HLS, cv2.COLOR_RGB2GRAY # 2 –∏–∑-–∑–∞ —Ç–æ–≥–æ —á—Ç–æ two —Å–æ–∑–≤—É—á–Ω–æ —Å to
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
""" 
–ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
----------
- src (source) : –∫–∞—Ä—Ç–∏–Ω–∫–∞ (—Ç–µ–Ω–∑–æ—Ä) —É –∫–æ—Ç–æ—Ä–æ–π —Ç—Ä–µ–±—É–µ—Ç—Å—è –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–∞–Ω–∞–ª—ã
- code : –∫–æ–¥ –∫–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏–∏ (–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ: 4 - cv2.COLOR_BGR2RGB —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∞ —Ä–∞–≤–Ω–∞—è —á–µ—Ç—ã—Ä–µ–º)
"""

plt.imshow(img)
```

#### PIL.Image

### 3.1 –ö–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—è (np.concatenate) <a id='h31'>
    
[—Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ](#b) | [–≥–ª–∞–≤–∞](#h3)

–ü–µ—Ä–≤—ã–π –ø–æ–ª–µ–∑–Ω—ã–π –Ω–∞–≤—ã–∫ –ø—Ä–∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–µ –¥–∞–Ω–Ω—ã—Ö: –∫–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—è –∫–∞—Ä—Ç–∏–Ω–æ–∫. –ö–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—è - —ç—Ç–æ –æ–ø–µ—Ä–∞—Ü–∏—è —Å–∫–ª–µ–∏–≤–∞–Ω–∏—è (—Å—Ç—Ä–æ–∫, –∫–∞—Ä—Ç–∏–Ω–æ–∫, –≤–µ–∫—Ç–æ—Ä–æ–≤). –î–∞–≤–∞–π—Ç–µ —Å–Ω–∞—á–∞–ª–∞ –ø–æ—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –Ω–∞ –≤–µ–∫—Ç–æ—Ä–∞—Ö, –∫–æ–Ω–∫–∞—Ç–µ–Ω–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä–∞ `a`, `b` –∏ `c` —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ `np.concatenate()`. –ü–µ—Ä–≤—ã–π –µ–µ –ø–∞—Ä–∞–º–µ—Ç—Ä - —Å–ø–∏—Å–æ–∫ –≤–µ–∫—Ç–æ—Ä–æ–≤/–º–∞—Ç—Ä–∏—Ü/—Ç–µ–Ω–∑–æ—Ä–æ–≤ –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞–¥–æ —Å–∫–ª–µ–∏—Ç—å.

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7,8])
c = np.array([9, 10, 11, 12, 13, 14, 15])

np.concatenate([a, b, c])
```

[—Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ](#b) | [–≥–ª–∞–≤–∞](#h3)

–ò–Ω–æ–≥–¥–∞ –Ω–∞–º —Ç—Ä–µ–±—É–µ—Ç—Å—è —Å–æ–µ–¥–∏–Ω–∏—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π, —Å–æ–∑–¥–∞–≤ –Ω–æ–≤–æ–µ –∏–∑–º–µ—Ä–µ–Ω–∏–µ. –ü—Ä–æ—â–µ –ø–æ–Ω—è—Ç—å –µ—Å–ª–∏ —Å—Ä–∞–∑—É –ø–µ—Ä–µ–π—Ç–∏ –∫ –ø—Ä–∏–º–µ—Ä—É: –ø—É—Å—Ç—å —É –Ω–∞—Å –µ—Å—Ç—å —Ç—Ä–∏ —Å—Ç—Ä–æ–∫–∏ –º–∞—Ç—Ä–∏—Ü—ã, —Ö—Ä–∞–Ω—è—â–∏–µ—Å—è –≤ —Ä–∞–∑–Ω—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö

### 3.3 –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–¥—É—à–∫—É (np.pad) <a id='h33'>
    
[—Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ](#b) | [–≥–ª–∞–≤–∞](#h3)

–ü–æ–¥—É—à–∫–æ–π (pad) –Ω–∞–∑—ã–≤–∞—é—Ç –∑–Ω–∞—á–µ–Ω–∏—è, –æ–±—Ä–∞–º–ª—è—é—â–∏–µ –≤–µ–∫—Ç–æ—Ä/–º–∞—Ç—Ä–∏—Ü—É/—Ç–µ–Ω–∑–æ—Ä —Å –¥–≤—É—Ö —Å—Ç–æ—Ä–æ–Ω –∫–∞–∫–æ–≥–æ-–ª–∏–±–æ –∏–∑–º–µ—Ä–µ–Ω–∏—è. –ß—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å –æ —á–µ–º —Ä–µ—á—å, –æ–ø—è—Ç—å –∂–µ —Å—Ä–∞–∑—É –ø–µ—Ä–µ–π–¥–µ–º –∫ –ø—Ä–∏–º–µ—Ä—É

```python
a = np.array([1,2,3,4])

print(a)
```

–î–æ–±–∞–≤–∏–º –∫ —ç—Ç–æ–º—É –º–∞—Å—Å–∏–≤—É –ø–æ–¥—É—à–∫—É –∏–∑ –Ω—É–ª–µ–π —Ç–æ–ª—â–∏–Ω–æ–π 3 —Å –∫–∞–∂–¥–æ–π —Å—Ç–æ—Ä–æ–Ω—ã

```python
a_with_pad = np.pad(a, pad_width=3, constant_values=0)
"""
–ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
---------
array - numpy.ndarray
pad_width - —Ç–æ–ª—â–∏–Ω–∞ –ø–æ–¥—É—à–∫–∏
mode - —Ä–µ–∂–∏–º ('constant', 'edge', 'maximum', 'mean') 
constant_values - —ç—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä –∑–∞–¥–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –ø—Ä–∏ mode='constant' –∏ —É–∫–∞–∑—ã–≤–∞–µ—Ç –∫–∞–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø–æ–¥—É—à–∫—É
""" 

a_with_pad
```

–ï—Å–ª–∏ –º—ã —Ö–æ—Ç–∏–º —Å–ª–µ–≤–∞ –¥–æ–±–∞–≤–∏—Ç—å 3 –∑–Ω–∞—á–µ–Ω–∏—è, –∞ —Å–ø—Ä–∞–≤–∞ 1, —Ç–æ–≥–¥–∞ –≤ `pad_width` –º–æ–∂–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç—å –ø–∞—Ä—É –∑–Ω–∞—á–µ–Ω–∏–π

```python
np.pad(a, pad_width=(3, 1), constant_values=0)
```

–¢–µ–ø–µ—Ä—å –¥–∞–≤–∞–π—Ç–µ –¥–æ–±–∞–≤–∏–º –ø–æ–¥—É—à–∫—É –∫ –º–∞—Ç—Ä–∏—Ü–µ. –í –º–∞—Ç—Ä–∏—Ü–µ –º—ã –º–æ–∂–µ–º –¥–æ–±–∞–≤–∏—Ç—å –ø–æ–¥—É—à–∫—É –∫–∞–∫ –ø–æ –±–æ–∫–∞–º, —Ç–∞–∫ –∏ —Å–≤–µ—Ä—Ö—É/—Å–Ω–∏–∑—É. –ü–æ—ç—Ç–æ–º—É –ø—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å –º–∞—Ç—Ä–∏—Ü–µ–π –º—ã –º–æ–∂–µ–º –ø–µ—Ä–µ–¥–∞—Ç—å –¥–≤–µ –ø–∞—Ä—ã –∑–Ω–∞—á–µ–Ω–∏–π, –∑–∞–¥–∞—é—â–∏—Ö —Ç–æ–ª—â–∏–Ω—É: (—Å–≤–µ—Ä—Ö—É, —Å–Ω–∏–∑—É), (—Å–ª–µ–≤–∞, —Å–ø—Ä–∞–≤–∞) . –î–æ–±–∞–≤–∏–º –ø–æ–¥—É—à–∫—É —Ç–æ–ª—â–∏–Ω–æ–π 1 —Å–≤–µ—Ä—Ö—É, 2 —Å–Ω–∏–∑—É, 3 —Å–ª–µ–≤–∞ –∏ 4 —Å–ø—Ä–∞–≤–∞
